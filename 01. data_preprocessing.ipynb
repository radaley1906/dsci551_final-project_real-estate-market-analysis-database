# Database dataframes

## Home value Index table

```python
# Metro level
# ZHVI - A measure of the typical home value and market changes across a given region and housing type. 

# SFH - Zillow Home Value Index (ZHVI)
homevalue_zhvi = pd.read_csv('../data/home_value/Metro_zhvi_bdrmcnt_5_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv')
```

```python
# Zillow Home Value Index (ZHVI) table

# get only date columns between start and end
date_columns = [col for col in homevalue_zhvi.columns if start <= col <= end ]

# include regional data
region_columns = ['RegionID','RegionName','StateName']
selected_columns = region_columns + date_columns

# create dataframe
homevalue_zhvi_selected = homevalue_zhvi[selected_columns]

# update dataframe columns

# create city and state
homevalue_zhvi_selected = homevalue_zhvi_selected.copy()
homevalue_zhvi_selected['City'] = homevalue_zhvi_selected['RegionName'].str.split(",").str[0]
homevalue_zhvi_selected['State'] = homevalue_zhvi_selected['RegionName'].str.split(",").str[1].str.strip()

# drop extra columns
homevalue_zhvi_selected = homevalue_zhvi_selected.drop(columns=['RegionName', 'StateName'])

# create table
homevalue_zhvi_table = homevalue_zhvi_selected.copy()
homevalue_zhvi_table.head()
```

```python
df1_to_table = homevalue_zhvi_table
```

# Import dataframe into postgreSQL database

```python
# libraries to import data

import pandas as pd


import psycopg2 as pg2
from psycopg2 import sql
from psycopg2.extras import execute_values
```

# tables for database
- df1_to_table
- df2_to_table
- df3_to_table
- df4_to_table

```python
# Establish a connection to PostgreSQL database
mydb = 'test'
secret = 'password'

# pass in the database name 'dvdrental'
conn = pg2.connect(database= mydb,user='postgres', password = secret )

cur = conn.cursor()

mycursor = cur
```

## df1_to_table: home value index table

```python
# map pandas dype to Postgres types

type_map = {
    "int64": "BIGINT",
    "float64": "DOUBLE PRECISION",
    "object": "TEXT",
    "datetime64[ns]": "TIMESTAMP"
}

columns = []

for col, dtype in zip(df1_to_table.columns, df1_to_table.dtypes):
    pg_type = type_map.get(str(dtype), "TEXT")
    columns.append(
        sql.SQL("{} {}").format(
            sql.Identifier(col),
            sql.SQL(pg_type)
        )
    )

create_table = sql.SQL("CREATE TABLE {} ({})").format(
    sql.Identifier("home_value_index"),
    sql.SQL(", ").join(columns)
)

cur.execute(create_table)

# Bulk insert into table

insert_stmt = sql.SQL("INSERT INTO {} ({}) VALUES %s").format(
    sql.Identifier("home_value_index"),
    sql.SQL(" ,").join(map(sql.Identifier, df1_to_table.columns))

)

# convert Nan to Null

data = [
    tuple(None if pd.isna(val) else val for val in row)
    for row in df1_to_table.itertuples(index = False, name = None)

]

execute_values(cur, insert_stmt, data, page_size=500)

# commit everything
conn.commit()
cur.close()
conn.close()
```
